Importing libraries and loading data files.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder

```

```python
movies_data = pd.read_csv(r'C:\Users\winra\Documents\EECS-731-Project-3\data\external\movies.csv')
ratings_data = pd.read_csv(r'C:\Users\winra\Documents\EECS-731-Project-3\data\external\ratings.csv')
tags_data = pd.read_csv(r'C:\Users\winra\Documents\EECS-731-Project-3\data\external\tags.csv')
```

Changing type of tag column in tags_data to string

```python
tags_data["tag"]= tags_data["tag"].astype(str) 
```

Droping userId and timestamp from tags_data as they are not significant in clustering movies.

```python
tags_data= tags_data.drop(['userId', 'timestamp'], axis='columns')

#groupby('movieId').agg({'tag':'sum'}).reset_index()
```

Grouping the tags of movies on the basis of movie.

```python
agg_tags_data = tags_data.groupby('movieId')['tag'].agg(", ".join).reset_index()
#tags_data=tags_data.groupby(['movieId'])['tag'].transform(lambda x: ', '.join(x)).reset_index()
```

```python
#tags_data['movieId']= data[:,0:0]
#tags_data
```

Droping userId and timestamp from ratings_data as they are not significant.

```python
ratings_data=ratings_data.drop(['userId','timestamp'], axis ='columns')
```

Getting average ratings for movies

```python
ratings_data = ratings_data.groupby('movieId', as_index=False)['rating'].mean()
```

Merging dataframes movies_data & ratings_data on the basis of movieId

```python
all_data = pd.merge(movies_data, ratings_data, on=["movieId"])
#all_data.head()
```

Merging dataframes all_data & agg_tags_data on the basis of movieId

```python
all_data = pd.merge(all_data, agg_tags_data, on=["movieId"])
#all_data.head()
```

dropping title and tags to cluster the movies on the basis of rating and genres

```python
data= all_data.drop(['title','tag'], axis ='columns')
#kmeans = KMeans(n_clusters=95).fit(all_data['rating'].reshape(-1, 1))

```

encoding the genres column to fit the cluster with numerical values

```python
le=LabelEncoder()
genres=le.fit_transform(data["genres"])
```

```python
data['genres'] = genres
```

Initializing KMeans clustering model with 20 n_clusters model

```python
kmeans = KMeans(n_clusters=20,random_state=1)
```

Fitting clustering modelwith genres and rating 

```python
kmeans = kmeans.fit(data[['genres','rating']])
```

getting the centers

```python
centers = kmeans.cluster_centers_
```

Plotting the Cluster.

```python
plt.scatter(data['genres'], data['rating'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=50)
plt.show()
```

Saving the resulting data to CSV file.

```python
data.to_csv(r"C:\Users\winra\Documents\EECS-731-Project-3\data\external\processed_data.csv",index=False) 
```
